```python
from docx import Document
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd

%matplotlib inline
```

```python
document = Document('./3920.docx')
```

```python
parrafos = document.paragraphs
len(parrafos)
```

```python
conpes_texto = ""
```

```python
conpes_texto = ""    
for p in parrafos:
    conpes_texto = conpes_texto + "\n" + p.text
    
stop_words = set(stopwords.words('spanish'))

stop_words.add("(")

stop_words.add(")")

stop_words.add(",")

stop_words.add(".")

stop_words.add(":")

stop_words.add("%")

stop_words.add("así")

stop_words.add("La")

stop_words.add("por")

word_tokens = word_tokenize(conpes_texto)

filtered_conpes = []

for w in word_tokens:
    if w not in stop_words:
        filtered_conpes.append(w)

words = pd.Series(filtered_conpes)
words.value_counts().to_frame().head(10)
```

```python
def word_could(parrafos, palabras_excluidas = []):
    
    conpes_texto = ""    
    for p in parrafos:
        conpes_texto = conpes_texto + "\n" + p.text.lower()
        
    word_tokens = word_tokenize(conpes_texto)

    filtered_conpes = []
    
    stop_words_palabras_excluidas = set(stop_words).union(set(palabras_excluidas))

    for w in word_tokens:
        if w not in stop_words_palabras_excluidas:
            filtered_conpes.append(w)

    words = pd.Series(filtered_conpes)
    
    wordcloud = WordCloud(width = 1000, height = 800, 
                background_color ='white',
                min_font_size = 10).generate(" ".join(words)) 
  
    # plot the WordCloud image                        
    plt.figure(figsize = (10, 8), facecolor = None) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 

    plt.show() 
```

```python
word_could(parrafos)
```

```python
word_could(parrafos[:200], palabras_excluidas=["dato","datos","política","pública","público"])
```

```python
word_could(parrafos[400:600], palabras_excluidas=["dato","datos","política","pública","público"])
```

```python
conteo = []
for p in parrafos:
    conteo.append(p.text.lower().count("acceso"))
bins = 25
conteo_grafica = []
for i in range(bins):
    fondo = i*(len(conteo)//bins)
    tope = ((i+1)*(len(conteo)//bins)) if (((i+1)*(len(conteo)//bins)) < len(conteo)) else (len(conteo)-1)
    conteo_grafica.append(sum(conteo[fondo:tope]))
```

```python
pd.Series(conteo_grafica).plot()
```
